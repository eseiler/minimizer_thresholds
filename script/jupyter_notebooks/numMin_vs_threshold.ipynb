```python
def x_y(counts):
    leading_zeros = 0
    trailing_zeros = 0
    assert(any(counts) != 0)
    i = 0
    while (counts[i] == 0):
        leading_zeros += 1
        i += 1
    i = len(counts) - 1
    while (counts[i] == 0):
        trailing_zeros += 1
        i -= 1

    new_counts = counts[leading_zeros:len(counts) - trailing_zeros]
    indices = list(range(leading_zeros, len(counts) - trailing_zeros))
    assert(len(new_counts) == len(indices))
    print(indices, new_counts)
    return (indices, new_counts)
```
